Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
import folium
from folium.plugins import MarkerCluster

# Assuming you have loaded your data into a DataFrame called launch_records
# Replace launch_records with the name of your DataFrame if it's different

# Create a map centered at a specific location
map_center = (28.573255, -80.646895)  # For example, Kennedy Space Center
site_map = folium.Map(location=map_center, zoom_start=4)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Iterate over each launch record in the DataFrame
for index, record in spacex_df.iterrows():
    # Determine the marker color based on the launch outcome
    marker_color = 'green' if record['class'] == 1 else 'red'  # Assuming the column name is 'class'
    
    # Create a marker with the determined color and add it to the MarkerCluster
    folium.Marker(location=(record['Lat'], record['Long']), icon=folium.Icon(color=marker_color)).add_to(marker_cluster)

# Add the MarkerCluster to the map
marker_cluster.add_to(site_map)

# Display the map
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
from folium.plugins import MousePosition

# Define a formatter function for formatting latitude and longitude values
formatter = "function(num) {return L.Util.formatNum(num, 5);};"

# Create a MousePosition object
mouse_position = MousePosition(
    position='topright',  # Position of the mouse position control on the map
    separator=' Long: ',  # Separator between latitude and longitude values
    empty_string='NaN',   # String to display if no coordinates are available
    lng_first=False,      # Whether to display longitude first
    num_digits=20,        # Number of digits after the decimal point
    prefix='Lat:',        # Prefix for latitude value
    lat_formatter=formatter,  # Latitude formatter function
    lng_formatter=formatter   # Longitude formatter function
)

# Add the MousePosition control to the map
site_map.add_child(mouse_position)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
